```python
# Veri setlerini okuyoruz
df = pd.read_csv('/Users/canergiden/Desktop/MAC/AÄŸ Programlama /Veri Seti/heart-disease.csv')
```

```python
# Ã–zellikleri Ã¶lÃ§eklendiriyoruz
scaler = StandardScaler()
features = scaler.fit_transform(df.drop('target', axis=1))
```

```python
# NetworkX grafiÄŸi oluÅŸturuyoruz
G = nx.Graph()
```

```python
# Her bir dÃ¼ÄŸÃ¼m iÃ§in
for idx, row in df.iterrows():
    # DÃ¼ÄŸÃ¼mÃ¼ ekliyoruz
    G.add_node(idx, **{f'f{i}': val for i, val in enumerate(features[idx])})
    # SÄ±nÄ±fÄ± ekliyoruz
    G.nodes[idx]['label'] = row['target']
```

```python
# KenarlarÄ± ekliyoruz
for i in range(len(df)):
    for j in range(i+1, len(df)):
        G.add_edge(i, j)
```

```python
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)

# Node ve edge renklerini deÄŸiÅŸtir
nx.draw(G, pos, with_labels=False, node_size=50, node_color='lightblue', edge_color='darkgray', alpha=0.7)

# Node'larÄ±n kenarlarÄ±nÄ± kalÄ±nlaÅŸtÄ±r
nx.draw_networkx_edges(G, pos, width=0.5)

plt.title("Heart Disease Dataset")
plt.show()
```

```python
# DÃ¼ÄŸÃ¼m, kenar ve sÄ±nÄ±f bilgilerini iÃ§eren yeni bir veri seti oluÅŸturuyoruz
node_data = pd.DataFrame.from_dict(dict(G.nodes(data=True)), orient='index')
edge_data = pd.DataFrame(G.edges, columns=['source', 'target'])
data = node_data.merge(edge_data, how='outer', left_index=True, right_on='source').drop('source', axis=1)
```

```python
from torch_geometric.data import Data

# PyTorch Geometric veri yapÄ±sÄ±na dÃ¶nÃ¼ÅŸtÃ¼rÃ¼yoruz
x = torch.tensor(features, dtype=torch.float32)
y = torch.tensor(df['target'].values, dtype=torch.long)
edge_index = torch.tensor(list(G.edges)).t().contiguous()

data = Data(x=x, y=y, edge_index=edge_index)

print(data)
```

```python
# Model iÃ§in bazÄ± parametreleri tanÄ±mlÄ±yoruz
num_features = len(df.columns) - 1
num_classes = len(df['target'].unique())
num_nodes = len(df)
num_edges = num_nodes * (num_nodes - 1) // 2
```

```python
# Model tanÄ±mÄ±nÄ± yapalÄ±m
class GCN(torch.nn.Module):
    def __init__(self, hidden_channels):
        super().__init__()
        torch.manual_seed(1234567)
        self.conv1 = GCNConv(num_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, num_classes)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = x.relu()
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        return x
```

```python
# Train, validation ve test setleri oluÅŸturalÄ±m
train_size = int(0.8 * len(data.y))
val_size = int(0.1 * len(data.y))
test_size = len(data.y) - train_size - val_size

data.train_mask = torch.zeros(len(data.y), dtype=torch.bool)
data.train_mask[:train_size] = 1

data.val_mask = torch.zeros(len(data.y), dtype=torch.bool)
data.val_mask[train_size:train_size + val_size] = 1

data.test_mask = torch.zeros(len(data.y), dtype=torch.bool)
data.test_mask[train_size + val_size:] = 1
```

```python
# Modeli oluÅŸturalÄ±m ve eÄŸitelim
model = GCN(hidden_channels=16)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

def train():
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss

def test(mask):
    model.eval()
    out = model(data.x, data.edge_index)
    pred = out.argmax(dim=1)
    correct = pred[mask] == data.y[mask]
    acc = int(correct.sum()) / int(mask.sum())
    return acc

for epoch in range(1, 101):
    loss = train()
    val_acc = test(data.val_mask)
    test_acc = test(data.test_mask)
    print(f'Epoch: {epoch:03d}, Loss: {loss:.4f}, Val: {val_acc:.4f}, Test: {test_acc:.4f}')
```

```python
# Modeli deÄŸerlendirelim
model.eval()
out = model(data.x, data.edge_index)
```

```python
# Modelin doÄŸruluÄŸunu hesaplayalÄ±m
def calculate_accuracy(mask):
    model.eval()
    with torch.no_grad():
        logits = model(data.x, data.edge_index)
        pred = logits.argmax(dim=1)
        correct = pred[mask] == data.y[mask]
        acc = correct.sum().item() / mask.sum().item()
    return acc

train_acc = calculate_accuracy(data.train_mask)
val_acc = calculate_accuracy(data.val_mask)
test_acc = calculate_accuracy(data.test_mask)

print(f"Train Accuracy: {train_acc:.4f}")
print(f"Validation Accuracy: {val_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_simple_graph(model, data):
    model.eval()
    with torch.no_grad():
        out = model(data.x, data.edge_index)
        pred = out.argmax(dim=1)

        G = nx.Graph()
        for i in range(data.num_nodes):
            G.add_node(i, label=data.y[i].item(), predicted_label=pred[i].item())

        for i in range(data.edge_index.shape[1]):
            src = data.edge_index[0, i].item()
            tgt = data.edge_index[1, i].item()
            G.add_edge(src, tgt)

        plt.figure(figsize=(10, 8))
        pos = nx.spring_layout(G, seed=42)
        nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')
        nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
        labels = {i: f"{G.nodes[i]['label']}({G.nodes[i]['predicted_label']})" for i in G.nodes()}
        nx.draw_networkx_labels(G, pos, labels=labels, font_size=10, font_color='black', font_weight='bold')
        plt.title("Graph Visualization\n(Labels: Actual(Predicted))")
        plt.axis('off')
        plt.show()

visualize_simple_graph(model, data)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def visualize_graph_with_predictions(model, data):
    model.eval()
    with torch.no_grad():
        out = model(data.x, data.edge_index)
        pred = out.argmax(dim=1)

        G = nx.Graph()
        for i in range(data.num_nodes):
            G.add_node(i, label=data.y[i].item(), predicted_label=pred[i].item(), correct=data.y[i].item() == pred[i].item())

        for i in range(data.edge_index.shape[1]):
            src = data.edge_index[0, i].item()
            tgt = data.edge_index[1, i].item()
            G.add_edge(src, tgt)

        plt.figure(figsize=(12, 8))
        pos = nx.spring_layout(G, seed=42)
        node_colors = ['lightblue' if G.nodes[i]['correct'] else 'blue' for i in G.nodes()]
        nx.draw(G, pos, with_labels=False, node_size=50, node_color=node_colors, edge_color='gray')
        plt.title("Heart Disease Dataset with Predictions")
        plt.show()

visualize_graph_with_predictions(model, data)
```
